# 인스타그램 크롤링 이미지 다운로드 beautifulsoup selenium 사용법

```python
from urllib.request import urlopen
from urllib.parse import quote_plus
from bs4 import BeautifulSoup
from selenium import webdriver 
import time 

#https://www.instagram.com/explore/tags/

baseurl= 'https://www.instagram.com/explore/tags/'
plusurl=input('검색할 태그를 입력하세요:')
url = baseurl + quote_plus(plusurl)

driver = webdriver.Chrome()
driver.get(url)
time.sleep(3)
#time 을 써야하는 이유 : 속도를 맞춰주기 위해서 
#selenium 이 느리기 때문에 버튼을 누르고 기다리는 시간을  벌어주기위해
html=driver.page_source
soup=BeautifulSoup(html)

insta=soup.select('.v1Nh3.kIKUG._bz0w')
#print(insta[0])

n=1
for i in insta:
    print('https://www.instagram.com'+i.a['href'])
    #img2 폴더에 저장하기
    imgurl=i.select_one('.KL4Bh').img['src'] #KL4Bh 라는 class에서 가져오는데, 그 중 img 에서 src만
    with urlopen(imgurl) as f:
        with open('./img2/' + plusurl + str(n) + '.jpg', 'wb') as h:
            img= f.read()
            h.write(img)
    n += 1
    print(imgurl)
    print()

driver.close()
```
