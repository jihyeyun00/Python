```python
#원핫인코딩:
num_classes=10
y_train=tf.keras.utils.to_categorical(y_train,num_classes)
y_val=tf.keras.utils.to_categorical(y_val,num_classes)
y_test=tf.keras.utils.to_categorical(y_val,num_classes)
```

```python
x=tf.placeholder(tf.float32,[None,784])
y=tf.placeholder(tf.float32,[None,10])
```

```python
def mlp(x):
    # hidden layer1
    w1 = tf.Variable(tf.random_uniform([784,256]))
    b1 = tf.Variable(tf.zeros([256]))
    h1 = tf.nn.relu(tf.matmul(x, w1) + b1)
    # hidden layer2
    w2 = tf.Variable(tf.random_uniform([256,128]))
    b2 = tf.Variable(tf.zeros([128]))
    h2 = tf.nn.relu(tf.matmul(h1, w2) + b2)
    # output layer
    w3 = tf.Variable(tf.random_uniform([128,10]))
    b3 = tf.Variable(tf.zeros([10]))
    logits= tf.matmul(h2, w3) + b3
    
    return logits
```

```python
logits=mlp(x)
```

```python
loss_op = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=logits, labels=y))
train_op=tf.train.AdamOptimizer(learning_rate=0.01).minimize(loss_op)
```

```python
#그래프 돌리기
init=tf.global_variables_initializer()
epoch_cnt=30
batch_size=1000
iteration=len(x_train)//batch_size
```

```python
# with tf.Session() as sess:
    sess.run(init)
    for epoch in range(epoch_cnt):
        avg_loss=0.
        start=0
        end=batch_size
        
        for i in range(iteration):
            _,loss=sess.run([train_op,loss_op],
                           feed_dict={x:x_train[start:end],y:y_train[start:end]})
            start +=batch_size
            end +=batch_size
            avg_loss +=loss/iteration
            
        preds=tf.nn.sotfmax(logits)
        correct_prediction=tf.equal(tf.agrmax(preds,1),tf.argmax(y,1))
        accuracy=tf.reduce_mean(tf.cast(correct_prediction,'float'))
        cur_val_acc=accuracy.eval({x:x_val,y:y_val})
        print('epoch:' +str(epoch)+ ',validation accuracy:' +str(cur_val_acc)+ ',loss:'+ str(avg_loss))
      
    preds=tf.nn.softmax(logits)
    correct_prediction=tf.equal(tf.argmax(preds,1),tf.argmax(y,1))
    accuracy=tf.reduce_mean(tf.cast(correct_prediction,))
    print('[test accuracy]: ',accuracy.eval({x:x_test,y:y_test}))
```
