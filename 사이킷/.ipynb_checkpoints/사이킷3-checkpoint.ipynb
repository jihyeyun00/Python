```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np


#이미지를 파일로 출력하고 로딩한다음 글씨만 추출
def make_hello(N=1000, rseed=42):
    fig, ax = plt.subplots(figsize=(4, 1))
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1)
    ax.axis('off')
    ax.text(0.5, 0.4, 'HELLO', va='center', ha='center', weight='bold', size=85)
    fig.savefig('hello.png')
    plt.close(fig)   

    from matplotlib.image import imread
    data = imread('hello.png')[::-1, :, 0].T   
    print("이미지차원", data.shape)
   
    print(data)
    rng = np.random.RandomState(rseed)
    X = rng.rand(4 * N, 2)
    print("만든 갯수",X.shape)
    print((X * data.shape).shape)
    i, j = (X * data.shape).astype(int).T
   
    mask = (data[i, j] < 1)
    X = X[mask]
    print("새로운X갯수", X.shape)
    print("원래이미지의 차수 ", data.shape)
    X[:, 0] *= (data.shape[0] / data.shape[1])
    X = X[:N]

    return X[np.argsort(X[:, 0])]
 
```

```python
X=make_hello(1000)
colorize=dict(c=X[:,0],cmap=plt.cm.get_cmap('rainbow',5))

plt.scatter(X[:,0],X[:,1],**colorize)
plt.axis('equal');
```

```python
import numpy as np

print(X.shape)
def rotate(X,angle):
    theta=np.deg2rad(angle) #사람이 식별할수 있는 각도를 라디안으로 ,라디안은 호의 길이
    R=[[np.cos(theta),np.sin(theta)],  #2X2 2차 행렬 
      [-np.sin(theta),np.cos(theta)]]
    print(type(R))
    return np.dot(X,R)   #1000X2 2x2=>1000X2
X2=rotate(X,20)+5
plt.scatter(X2[:,0],X2[:,1],**colorize)
plt.axis('equal')
```

```python
#각 점들간의 상호거리 
from sklearn.metrics import pairwise_distances #디폴트 -유틀리디안
D=pairwise_distances(X)  #거리행렬
print(D.shape)
D[:5,:5]
plt.imshow(D,zorder=2,cmap='Blues',interpolation='nearest')
plt.colorbar()
```

```python
D2=pairwise_distances(X2)
np.allclose(D,D2) #원형유지하고 있음
```

```python
import numpy as np
rng=np.random.RandomState(10) #seed값 고정
C=rng.randn(3,3) #normal 
print(np.dot(C,C.T)) #전치행렬 ,행렬의 거듭제곱
#정방행렬이면서 대칭행렬 나옴
e,V=np.linalg.eigh(np.dot(C,C.T)) #고유값분해
print("eigenvector",V) #고유벡터
print("eigenvalue",e) #고유값
np.dot(V[1],V[2]) #두벡터의 내적 =>직교
```

```python

def random_projection(X,dimension=3,rseed=42):  #random은 분석용으로는 못씀 시각화용
    assert dimension >= X.shape[1]   #행,열 (2차원에서 3차원으로 만듦) ,차원 확대하라
    rng=np.random.RandomState(rseed)
    C=rng.randn(dimension,dimension) #3X3 으로 데이터가 만들어짐
    print("C",C.shape)
    print(np.dot(C,C.T)) #정방행렬이면서 대칭행렬
    e,V=np.linalg.eigh(np.dot(C,C.T)) #고유치, 고유벡터
    print("V",V.shape)  #3x3 으로 나옴
    print('차원은',V[:X.shape[1]]) #3차원이 나옴
    return np.dot(X,V[:X.shape[1]])
print(X.shape)
print(X.shape[1])
print("데이터의 차원은",X.shape)
X3=random_projection(X,3)
X3.shape  #1000x3
    
```

```python
def make_hello_s_curve(X):
    t=(X[:,0]-2)*0.75*np.pi
    x=np.sin(t)
    y=X[:,1]
    z=np.sign(t)*(np.cos(t)-1) 
    return np.vstack((x,y,z)).T
XS=make_hello_s_curve(X)
ax=plt.axes(projection="3d")
ax.scatter3D(XS[:,0],XS[:,1],XS[:,2],**colorize)
#MDS 는 선형에는 강하지만 비선형에는 취약하다
```

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from matplotlib import pylab

from sklearn import manifold
from matplotlib import pylab
from sklearn.manifold import MDS
from mpl_toolkits import mplot3d
import os #디렉터리 관리

CHART_DIR="./"  
colors=['r','g','b']
markers=['o',6,'*']
def plot_iris_mds():
    iris = load_iris()
    X = iris.data
    y = iris.target
    fig = pylab.figure(figsize=(10, 4))
    ax = fig.add_subplot(121, projection='3d')
    ax.set_facecolor('white')
    mds = manifold.MDS(n_components=3) #아이리스 변수 4개인데 3개로 
    Xtrans = mds.fit_transform(X)   #150X3 아이리스 행 150 개 3개의 변수로 (차원축소해서 3)
    for cl, color, marker in zip(np.unique(y), colors, markers):
        ax.scatter(
            Xtrans[y == cl][:, 0], Xtrans[y == cl][:, 1], Xtrans[y == cl][:, 2], c=color, marker=marker, edgecolor='black')
    pylab.title("3차원에서 iris MDS")
    ax.view_init(10, -15)  #카메라 각도 조절

    mds = manifold.MDS(n_components=2)  #2차원으로 
    Xtrans = mds.fit_transform(X)

    ax = fig.add_subplot(122)
    for cl, color, marker in zip(np.unique(y), colors, markers): 
        ax.scatter(
            Xtrans[y == cl][:, 0], Xtrans[y == cl][:, 1], c=color, marker=marker, edgecolor='black')
    pylab.title("2차원에서 iris MDS")

    filename = "mds_demo_iris.png"
    pylab.savefig(os.path.join(CHART_DIR, filename), bbox_inches="tight") 
print(plot_iris_mds())
```

```python
from sklearn.svm import SVC  #Classificition 분류
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler  

#kmeans,PCA는 전처리에 민감 
#preprocessing :전처리 ,전처리를 안해주면 사이즈가 큰 것이 

cancer=load_breast_cancer()
X_train,X_test,y_train,y_test=train_test_split(cancer.data,cancer.target,random_state=0)
scaler=MinMaxScaler().fit(X_train) #chaninng:MinMaxScaler().fit(X_train) :점으로 바로 다음함수 호출
X_train_scaled=scaler.transform(X_train)
svm=SVC()
svm.fit(X_train_scaled,y_train)
X_test_scaled=scaler.transform(X_test)
print('테스트점수:{:.2f}'.format(svm.score(X_test_scaled,y_test)))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={"C":[0.001,0.01,0.1,1,10,100],
           'gamma':[0.001,0.01,0.1,1,10,100]}
grid=GridSearchCV(SVC(),param_grid=param_grid,cv=5)
grid.fit(X_train_scaled,y_train)
print('최상의 교차 검증 정확도:{:.2f}'.format(grid.best_score_))
print('테스트 점수:{:.2f}'.format(grid.score(X_test_scaled,y_test)))
print('최적의 매개변수:',grid.best_params_)
```

```python
from sklearn.pipeline import Pipeline
#파이프이름 쓰는 이유: 참조:파라미터 전달하기위해
pipe=Pipeline([('scaler',MinMaxScaler()),("svm",SVC())])
pipe.fit(X_train,y_train)
print('테스트점수:{:.2f}'.format(pipe.score(X_test,y_test)))
```

```python
#pipeline +GridSearchCV 는 다양한 테스트 문제를 해결
#regularization 규제: 과적합 C 가 강도
param_grid={"svm__C":[0.001,0.01,0.1,1,10,100], #대충 스팩트럼을 넓게 해서 주고
           'svm__gamma':[0.001,0.01,0.1,1,10,100]}

grid=GridSearchCV(pipe,param_grid=param_grid,cv=5) 
grid.fit(X_train_scaled,y_train)
print('최상의 교차 검증 정확도:{:.2f}'.format(grid.best_score_))
print('테스트 점수:{:.2f}'.format(grid.score(X_test_scaled,y_test)))
print('최적의 매개변수:',grid.best_params_)
```

```python
n_colors=10

#1바이트 =8비트 표현종류=>256 
china=load_sample_image("flower.jpg") #RGB(256 컬러 :1바이트)
#컬러값 정규화 0~1 정규화
china=np.array(china,dtype=np.float64)/255
```

```python
w,h,d=original_shape=tuple(china.shape)  #이미지 사이즈가 행x열, 3
print(w,h,d)  #427 , 640, 3 
assert d==3    #RGB만 들어와라 (흑백은 )
image_array=np.reshape(china,(w*h,d)) #계산하기 쉽게 학기 위해 2차원으로 변환
```

```python
#1000개의 행
image_array_sample=shuffle(image_array,random_state=0)[:1000]
#n_colors=64컬러로 군집화
kmeans=KMeans(n_clusters=n_colors,random_state=0).fit(image_array_sample)  #중심값 결정 64
```

```python
labels=kmeans.predict(image_array)  #라벨 427X640
```

```python
def recreate_image(codebook,labels,w,h):  #64컬러값 ,labels:이미지 픽셀수만큼 
    d=codebook.shape[1]  #64개의 중심값 64x3 
    image=np.zeros((w,h,d)) #원래 이미지 사이즈 
    label_idx=0
    for i in range(w):
        for j in range(h):
            image[i][j]=codebook[labels[label_idx]]
            label_idx +=1
    return image
```

```python
kmeans=KMeans(n_clusters=1)  #중심=1개
kmeans.fit(X)
distances=kmeans.transform(X) #각데이터의 중심으로 부터 값으로 변환

#ravel() 1차원
#값을 보고 인덱스를 소트하라 값:argsort
#:: 2개가 있는 경우
#값을 기준으로 인덱싱을 하는 이유 :그 값을 기준으로 다른 것을 제어하고 싶어서
sorted_idx=np.argsort(distances.ravel())[::-1][:5] #내림차순으로 변경
```

```python
f.ax=plt.subplots(figsize=(7,5))
ax.set_title('Single Cluster')
ax.scatter(X[:,0],X[:,1],label="Points")
ax.scatter(kmeans.cluster_centers_[:,0],
          kmeans.cluster_centers_[:,1],
          label="Centroid",color='r')
ax.scatter(X[sorted_idx][:,0],
          X[sorted_idx][:,1],
          label='이상치',edgecolors="g"
          facecolors="none",s=100)
ax.legend(loc='best')
```

```python
%matplotlib inline
from sklearn.datasets import fetch_lfw_people  #사람얼굴 다운 받는 것
import matplotlib.pyplot as plt
people=fetch_lfw_people(min_faces_per_person=20,resize=0.7)  #resize 사이즈 줄여줌
image_shape=people.images[0].shape
print(image_shape)  #87(세로)*65(가로) , 이미지를 읽을 때 :가로 *세로, 행렬:행(세로)*열
fig.axes=plt.subplots(2,5,figsize=(15,8),  #2*5 =10장을 읽는 것 
                     subplot_kw={'xticks':(),'yticks':()})
for target,image, ax in zip (people.target,people.images,axes.ravel()):
    ax.imshow(image)
    ax.set_title(people.target_names[target])
```

```python
print('이미지 사이즈:{}'.format(people.images.shape))
print('클래스 개수:{}'.format(len(people.target_names)))
```

```python
mask = np.zeros(people.target.shape, dtype = np.bool)
for target in np.unique(people.target) :
    mask[np.where(people.target == target)[0][:50]] = 1
X_people = people.data[mask]
y_people = people.target[mask]
X_people = X_people/255.
```

```python
X_train,X_test,y_train,y_test=train_test_split(
X_people,y_people,stratify=y_people,random_state=0)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=100,whiten=True,random_state=0).fit(X_train)
X_train_pca=pca.transform(X_train)
X_test_pca=pca.transform(X_test)
```

```python
fig.axes=plt.subplots(3,5,figsize=(15,12),
                     subplot_kw={'xticks':(),'yticks':()})
for i,(component,ax) in enumerate(zip(pca.components_,axes.ravel())):
    ax.imshow(component.reshape(image_shape,cmap='viridis')
    ax.set_title('주성분{}'.format((i+1)))
```
